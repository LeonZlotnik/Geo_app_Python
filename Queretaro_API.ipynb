```python
ciudad = input("Escriba que ciudad desea buscar: ")
```

```python
keyword = input("Escriba que desea ")
```

```python
lat = input("Ingrese la latitud: ")
```

```python
lon = input("Ingrese la longitud: ")
```

```python
#ilat = int(lat)
#ilon = int(lon)

coordienadas = lat+','+lon
```

```python
import requests
import json
import pandas as pd
from config import api_key

target_coordinates = coordienadas
target_search = ciudad
target_radius = 4000
target_keyword = keyword

params = {
    'location': target_coordinates,
    'search': target_search,
    'radius': target_radius,
    'keyword': target_keyword,
    'key': api_key
}

url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

response = requests.get(url, params=params)
```

```python
place_data = response.json()

print(json.dumps(place_data,indent=4, sort_keys=True))
```
